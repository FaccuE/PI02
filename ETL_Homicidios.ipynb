```python
H_V = H_V.replace('SD', np.nan)
```

dejamos en blanco los valores NaN de las columnas ROL y VICTIMA pasamos a SIN IDENTIFICAR los NaN de la columna SEXO, cambiamos a 0 los NaN de la columna EDAD y cambiamos a NO FALLECIO los valores NaN de la columna FECHA_FALLECIMIENTO

```python
H_V['ROL'] = H_V['ROL'].fillna('')
H_V['VICTIMA'] = H_V['VICTIMA'].fillna('')
```

```python
H_V['SEXO'] = H_V['SEXO'].fillna('SIN IDENTIFICAR')
```

```python
H_V['EDAD'] = H_V['EDAD'].fillna(0)
```

```python
H_V['FECHA_FALLECIMIENTO'] = H_V['FECHA_FALLECIMIENTO'].fillna('NO FALLECIO')
```

```python
H_V['EDAD'] = H_V['EDAD'].astype(int)
```

```python
H_V = H_V.rename(columns={'ID_hecho': 'ID'})
```

Eliminamos la columna VICTIMA ya que se encuentra en el otro DF

```python
H_V = H_V.drop('VICTIMA', axis=1)
```

```python
H_V
```

```python
H_C = H_H.merge(H_V, on='ID', how='left')
```

```python
H_C
```

Eliminamos los espacios en la columna LUGAR_DEL_HECHO y eliminamos las comillas

```python
H_C['LUGAR_DEL_HECHO'] = H_C['LUGAR_DEL_HECHO'].str.strip()
```

```python
H_C['LUGAR_DEL_HECHO'] = H_C['LUGAR_DEL_HECHO'].str.replace('"', '')
```

```python
H_C
```

```python
H_C.info()
```

Cambiamos el formato y el tipo de columna de FECHA_FALLECIMIENTO

```python
H_C['FECHA_FALLECIMIENTO'] = pd.to_datetime(H_C['FECHA_FALLECIMIENTO'], errors='coerce')
H_C['FECHA_FALLECIMIENTO'] = H_C['FECHA_FALLECIMIENTO'].dt.strftime('%Y-%m-%d')
```

```python
H_C['FECHA_FALLECIMIENTO'] = H_C['FECHA_FALLECIMIENTO'].astype('datetime64[ns]')
```
